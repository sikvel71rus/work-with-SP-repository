```python
import pandas as pd
import datetime
from datetime import datetime, date
StartData = datetime.now().time()
print('Начало: ',datetime.now().time())
date1 = datetime.now().time()
xls_file = pd.ExcelFile('D:\\20180606_export_meta_inds.xlsx')
df = xls_file.parse('Export Worksheet')
xls_file2 = pd.ExcelFile('D:\\Period i izmeren.xlsx')
df2 = xls_file2.parse('Периодичность (2)')
df3 = xls_file2.parse('ЕдИзмер (2)')

df['Standart_Izmerenie'] = ''
for i in range (df.shape[0]):
    df['Standart_Izmerenie'][i] = str(sorted(str(df['Измерения'][i]).replace("/", "|").replace(" | ","|").replace("| ","|").replace("  |","|").split('|')))
    print('izmer',i)

#Создаём новые столбцы со стандартизированными единицами измерения и периодичностью
df['Standart_ED_Izmerenie'] = ''
df['Standart_Period'] = ''
for i in range(df.shape[0]):
    print('ed',i)
    for j in range (df2.shape[0]):
        if df['Периодичность'][i] == df2['old'][j]:
            df['Standart_Period'][i] = df2['new'][j]
            break
    for m in range (df3.shape[0]):
        if df['Ед. изм.'][i] == df3['old'][m]:
            df['Standart_ED_Izmerenie'][i] = df3['new'][m]
            break
            
#Создаю шапку для всех последующих таблиц
dfCap = df
for i in range (dfCap.shape[0]):
    dfCap = dfCap.drop(i)
    
    
dfFullDuplicates = dfCap
df['TableFullScep'] = ''
for i in range (df.shape[0]):
    print('dfFullDuplicates',i)
    df['TableFullScep'][i] = str(df['Код'][i]) + str(df['Наименование'][i]) + str(df['Standart_ED_Izmerenie'][i]) + str(df['Standart_Period'][i]) + str(df['Начало временного ряда'][i]) + str(df['Окончание временного ряда'][i]) + str(df['Крайняя дата передачи'][i]) + str(df['Кол-во записей в показателе'][i]) + str(df['Standart_Izmerenie'][i])
counts = df.groupby('TableFullScep').size()
dfDubli = pd.DataFrame(counts, columns = ['size'])
dfDubli = dfDubli[dfDubli['size'] > 1]
for i in range (df.shape[0]):
    for j in range (dfDubli.shape[0]):
        if (df['TableFullScep'][i]) == (dfDubli.index[j]):
            dfMiddle = df.iloc[i].copy()
            dfFullDuplicates = dfFullDuplicates.append(dfMiddle, ignore_index = True)
            dfMiddle = ""
            
dfNameDuplicates = dfCap
dfMiddle = ""
df['TableFullScep'] = ''
for i in range (df.shape[0]):
    print('dfNameDuplicates',i)
    df['TableFullScep'][i] = str(df['Наименование'][i])
counts = df.groupby('TableFullScep').size()
dfDubli = pd.DataFrame(counts, columns = ['size'])
dfDubli = dfDubli[dfDubli['size'] > 1]
for i in range (df.shape[0]):
    for j in range (dfDubli.shape[0]):
        if (df['TableFullScep'][i]) == (dfDubli.index[j]):
            dfMiddle = df.iloc[i].copy()
            dfNameDuplicates = dfNameDuplicates.append(dfMiddle, ignore_index = True)
            dfMiddle = ""

            
dfGroupDuplicates = dfCap
dfMiddle = ""
df['TableFullScep'] = ''
for i in range (df.shape[0]):
    print('dfGroupDuplicates',i)
    df['TableFullScep'][i] = str(df['Наименование'][i]) + str(df['Standart_ED_Izmerenie'][i]) + str(df['Standart_Period'][i]) + str(df['Standart_Izmerenie'][i])
counts = df.groupby('TableFullScep').size()
dfDubli = pd.DataFrame(counts, columns = ['size'])
dfDubli = dfDubli[dfDubli['size'] > 1]
for i in range (df.shape[0]):
    for j in range (dfDubli.shape[0]):
        if (df['TableFullScep'][i]) == (dfDubli.index[j]):
            dfMiddle = df.iloc[i].copy()
            dfGroupDuplicates = dfGroupDuplicates.append(dfMiddle, ignore_index = True)
            dfMiddle = ""
            
#Сохраняем результаты в файл excel
file_name = 'Дубликаты реестр показателей 2'
sheet_name = 'Реестр'
sheet_name1 = 'Дубликаты название'
sheet_name2 = 'Полные дубликаты'
sheet_name3 = 'Дубли Наим Измер ед Пер'
writer = pd.ExcelWriter('D:\\' + file_name + '.xlsx', engine = 'xlsxwriter')
df.to_excel(writer, sheet_name = sheet_name, index = False)
dfNameDuplicates.to_excel(writer, sheet_name = sheet_name1, index = False)
dfGroupDuplicates.to_excel(writer, sheet_name = sheet_name3, index = False)
dfFullDuplicates.to_excel(writer, sheet_name = sheet_name2, index = False)
writer.save()
print('Начало',StartData)
print('Конец: ',datetime.now().time())
    
```
